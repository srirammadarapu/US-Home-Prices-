```python
final_df = df_long.copy()

# Loop through other DataFrames (indexes 1 to 4)
for i, df1 in enumerate(others[1:]):
    print(f"Processing df at index {i+1}")

    if 'DATE' in df1.columns:
        df1['DATE'] = pd.to_datetime(df1['DATE'])
        df1['Year'] = df1['DATE'].dt.year
        df1.set_index('Year', inplace=True)
    elif 'Year' in df1.columns:
        df1.set_index('Year', inplace=True)
    else:
        print(f"Skipping df at index {i+1} — missing DATE/Year")
        continue

    # Merge with the main DataFrame on Year
    final_df = final_df.merge(df1, on='Year', how='left')
```
